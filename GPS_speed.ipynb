```python
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('TkAgg') # Need to use in order to run on mac
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import cnames
from matplotlib import animation

dataanim = pd.read_csv('/Users/Clauzinha/Dropbox/Neuro_Doctor/Project/Project gait_GPS/OpenSim/GIL06/edit_GIL06_free2_ik.csv') # only coordinate data

dataanim.columns  = np.repeat(range(32),3) # store cols like this for simplicity
N_tag = dataanim.shape[1]/3 # nr of tags used (all)

N_trajectories = N_tag

t = np.linspace(0,np.size(dataanim,0)) # pseudo time-vector for first walking activity
x_t = np.zeros(shape=(N_tag,dataanim.shape[0],3)) # empty animation array (3D)

for tag in range(32):
    # store data in numpy 3D array: (tag,time-stamp,xyz-coordinates)
    x_t[tag,:,:] = dataanim[tag]

x_t = x_t[:, :, [0, 2, 1]]

# Set up figure & 3D axis for animation
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1], projection='3d')
ax.axis('on')

# choose a different color for each trajectory
colors = plt.cm.jet(np.linspace(0, 1, N_trajectories))
# set up trajectory lines
lines = sum([ax.plot([], [], [], '-', c=c) for c in colors], [])
# set up points
pts = sum([ax.plot([], [], [], 'o', c=c) for c in colors], [])

ax.set_xlim3d([0,3])
ax.set_ylim3d([-1,1]) 
ax.set_zlim([0, 1]) 
ax.set_xlabel('X [mm]')
ax.set_ylabel('Y [mm]')
ax.set_zlabel('Z [mm]')

# set point-of-view: specified by (altitude degrees, azimuth degrees)
ax.view_init(30, 0)

# initialization function: plot the background of each frame
def init():
    for line, pt in zip(lines, pts):
        # trajectory lines
        line.set_data([], [])
        line.set_3d_properties([])
        # points
        pt.set_data([], [])
        pt.set_3d_properties([])
    return lines + pts

# animation function.  
def animate(i):
    i = (5 * i) % x_t.shape[1]

    for line, pt, xi in zip(lines, pts, x_t):
        x, y, z = xi[:i].T 
        pt.set_data(x[-1:], y[-1:])
        pt.set_3d_properties(z[-1:])

    ax.view_init(30, 0.3 * i)
    fig.canvas.draw()
    return lines + pts #+ stick_lines

# Animation
anim = animation.FuncAnimation(fig, animate, frames=500, interval=100, blit=True)

plt.show()
```

```python
# Open .txt files for Free and Fast speed
import numpy as np
gfastR = np.loadtxt('/Users/Clauzinha/Dropbox/Python/gaitfast_r.txt')
gfastL = np.loadtxt('/Users/Clauzinha/Dropbox/Python/gaitfast_l.txt')

gfreeR = np.loadtxt('/Users/Clauzinha/Dropbox/Python/gaitfree_r.txt')
gfreeL = np.loadtxt('/Users/Clauzinha/Dropbox/Python/gaitfree_l.txt')
```

```python
# Calculate RMS of each GVS
from numpy import mean, sqrt, square, arange
GVS_plv_tilt = sqrt(mean(square(gfastR[1:50] - gfreeR[1:50])))
GVS_plv_list = sqrt(mean(square(gfastR[51:100] - gfreeR[51:100])))
GVS_plv_rot = sqrt(mean(square(gfastR[101:150] - gfreeR[101:150])))
GVS_hip_flx_r = sqrt(mean(square(gfastR[151:200] - gfreeR[151:200])))
GVS_hip_add_r = sqrt(mean(square(gfastR[201:250] - gfreeR[201:250])))
GVS_hip_rot_r = sqrt(mean(square(gfastR[251:300] - gfreeR[251:300])))
GVS_knee_r = sqrt(mean(square(gfastR[301:350] - gfreeR[301:350])))
GVS_ankle_r = sqrt(mean(square(gfastR[351:400] - gfreeR[351:400])))
GVS_footp_r = sqrt(mean(square(gfastR[401:450] - gfreeR[401:450])))

GVS_hip_flx_l = sqrt(mean(square(gfastL[151:200] - gfreeL[151:200])))
GVS_hip_add_l = sqrt(mean(square(gfastL[201:250] - gfreeL[201:250])))
GVS_hip_rot_l = sqrt(mean(square(gfastL[251:300] - gfreeL[251:300])))
GVS_knee_l = sqrt(mean(square(gfastL[301:350] - gfreeL[301:350])))
GVS_ankle_l = sqrt(mean(square(gfastL[351:400] - gfreeL[351:400])))
GVS_footp_l = sqrt(mean(square(gfastL[401:450] - gfreeL[401:450])))
```
