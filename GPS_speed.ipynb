# GPS_speed study

Based on Baker et al.(2009), the aim of this study is to calculated the gait profile score (GPS) and gait variables score (GVS) of the gait in different walking speeds. The database is from Liu et al.(2008) https://simtk.org/home/mspeedwalksims.
Eight children were collected during walking overground at very slow, slow, free and fast speed.

```python
%matplotlib inline
from __future__ import division, print_function
import math
import numpy as np
import pandas as pd
from numpy import *
from numpy import mean, sqrt, square, arange
from matplotlib import pyplot as plt
from tnorm import tnorm
from IPython.display import HTML, display
from scipy import stats
```

```python
variable = ['plvtilt','plvlist','plvrot','hipflxR','hipaddR','hiprotR','kneeR','ankleR','footR','hipflxL','hipaddL','hiprotL','kneeL','ankleL','footL']
velocity = ['fast','free','slow','xslow']  # 0=fast 1=free 2=slow 3=xslow
subjects = ['01','02','03','04','06','08','011','012'] # 0=01 1=02 2=03 3=04 4=06 5=08 6=011 7=012
path = './dataOpenSim/GIL'
extension = '.mot'
```

```python
# Creating matrix to storage the new data normalized (every 2%). 480 = 15variables * 8 subjects * 4 velocities
# Reading files
dic = {}
for s, sub in enumerate(subjects):
    for v, vel in enumerate(velocity):
        filename = path + sub + '_' + vel + '_ik' + extension
        #print(filename)
        data = np.loadtxt(filename, skiprows=12, unpack=False)
        plvtilt, plvlist, plvrot = data[:,4], data[:,5], data[:,6]
        hipflxR, hipaddR, hiprotR = data[:,7], data[:,8], data[:,9]
        kneeR, ankleR = data[:,10], data[:,11]
        hipflxL, hipaddL, hiprotL = data[:,12], data[:,13], data[:,14]
        kneeL, ankleL = data[:,15], data[:,16]      
        anklejcR, toeR = data[:,78:81], data[:,81:84]
        anklejcL, toeL = data[:,114:117], data[:,117:120]
        fxR, fyR, fzR = data[:,246], data[:,247], data[:,248]
        fxL, fyL, fzL = data[:,252], data[:,253], data[:,254]       
        # Calculating foot progression
        footR = np.arctan2(toeR[:,2] - anklejcR[:,2], toeR[:,0] - anklejcR[:,0])*180/np.pi
        footL = np.arctan2(toeL[:,2] - anklejcL[:,2], toeL[:,0] - anklejcL[:,0])*180/np.pi      
        # Normalizing the data
        t = np.linspace(0,100,len(data))
        tn = np.linspace(0,100,101)
        plvtiltn = np.interp(tn,t,plvtilt)
        plvlistn = np.interp(tn,t,plvlist)
        plvrotn = np.interp(tn,t,plvrot)
        hipflxRn = np.interp(tn,t,hipflxR)
        hipaddRn = np.interp(tn,t,hipaddR)
        hiprotRn = np.interp(tn,t,hiprotR)
        kneeRn = np.interp(tn,t,kneeR)
        ankleRn = np.interp(tn,t,ankleR)
        hipflxLn = np.interp(tn,t,hipflxL)
        hipaddLn = np.interp(tn,t,hipaddL)
        hiprotLn = np.interp(tn,t,hiprotL)
        kneeLn = np.interp(tn,t,kneeL)
        ankleLn = np.interp(tn,t,ankleL)
        footRn = np.interp(tn,t,footR)
        footLn = np.interp(tn,t,footL)
        fxRn = np.interp(tn,t,fxR)
        fyRn = np.interp(tn,t,fyR)
        fzRn = np.interp(tn,t,fzR)
        fxLn = np.interp(tn,t,fxL)
        fyLn = np.interp(tn,t,fyL)
        fzLn = np.interp(tn,t,fzL)
        # Taking every 2% of the stance phase
        plvtilt2 = plvtiltn[::2,]
        plvlist2 = plvlistn[::2,]
        plvrot2 = plvrotn[::2,]  
        hipflxR2 = hipflxRn[::2,]
        hipaddR2 = hipaddRn[::2,]
        hiprotR2 = hiprotRn[::2,]  
        kneeR2 = kneeRn[::2,]
        ankleR2 = ankleRn[::2,]  
        hipflxL2 = hipflxLn[::2,]
        hipaddL2 = hipaddLn[::2,]
        hiprotL2 = hiprotLn[::2,]  
        kneeL2 = kneeLn[::2,]
        ankleL2 = ankleLn[::2,]  
        footR2 = footRn[::2,]
        footL2 = footLn[::2,]  
        fxR2 = fxRn[::2,]
        fyR2 = fyRn[::2,]
        fzR2 = fzRn[::2,] 
        fxL2 = fxLn[::2,]
        fyL2 = fyLn[::2,]
        fzL2 = fzLn[::2,] 
        # Changing the signal if the subject landed with the right foot
        if s==0 and v==0 or s==1 and v==0 or s==1 and v==1 or s==1 and v==2 or s==2 and v==1 or s==2 and v==2 or s==3 and v==0 or s==3 and v==2 or s==3 and v==3 or s==6 and v==0 or s==7 and v==0 or s==7 and v==2:
            footR2 = -footR2
            footL2 = -footL2
            plvtilt2 = -plvtilt2
            plvlist2 = -plvlist2
            plvrot2 = -plvrot2
        else:
            footR2 = footR2
            footL2 = footL2
            plvtilt2 = plvtilt2
            plvlist2 = plvlist2
            plvrot2 = plvrot2   
        
        # plot variables
        if 0: 
            hfig, hax = plt.subplots(4,3, sharex = True, squeeze=True, figsize=(10,10))
            hax[0,0].plot(plvtilt2, linewidth=2, alpha=.3)
            hax[0,0].set_ylabel('Pelvis Tilt[deg]', fontsize=12)
            hax[0,1].plot(plvlist2, linewidth=2, alpha=.3)
            hax[0,1].set_ylabel('Pelvis List[deg]', fontsize=12)
            hax[0,2].plot(plvrot2, linewidth=2, alpha=.3)
            hax[0,2].set_ylabel('Pelvis Rotation[deg]', fontsize=12)
            hax[1,0].plot(hipflxR2, linewidth=2, alpha=.3), hax[1,0].plot(hipflxL2, 'r', linewidth=2, alpha=.3) 
            hax[1,0].set_ylabel('Hip Flexion[deg]', fontsize=12)
            hax[1,1].plot(hipaddR2, linewidth=2, alpha=.3), hax[1,1].plot(hipaddL2, 'r', linewidth=2, alpha=.3)
            hax[1,1].set_ylabel('Hip Adduction[deg]', fontsize=12)
            hax[1,2].plot(hiprotR2, linewidth=2, alpha=.3), hax[1,2].plot(hiprotL2, 'r', linewidth=2, alpha=.3)
            hax[1,2].set_ylabel('Hip Rotation[deg]', fontsize=12)
            hax[2,0].plot(kneeR2, linewidth=2, alpha=.3), hax[2,0].plot(kneeL2, 'r', linewidth=2, alpha=.3)
            hax[2,0].set_ylabel('Knee Flexion[deg]', fontsize=12)
            hax[2,1].plot(ankleR2, linewidth=2, alpha=.3), hax[2,1].plot(ankleL2, 'r', linewidth=2, alpha=.3)
            hax[2,1].set_ylabel('Ankle DF[deg]', fontsize=12)
            hax[2,2].plot(footR2, linewidth=2, alpha=.3), hax[2,2].plot(footL2, 'r', linewidth=2, alpha=.3)
            hax[2,2].set_ylabel('Foot Prog[deg]', fontsize=12)
            hax[3,0].plot(fxR2, linewidth=2, alpha=.3), hax[3,0].plot(fxL2, 'r', linewidth=2, alpha=.3)
            hax[3,0].set_ylabel('FX[deg]', fontsize=12)
            hax[3,1].plot(fyR2, linewidth=2, alpha=.3), hax[3,1].plot(fyL2, 'r', linewidth=2, alpha=.3)
            hax[3,1].set_ylabel('FY[deg]', fontsize=12)
            hax[3,2].plot(fzR2, linewidth=2, alpha=.3), hax[3,2].plot(fzL2, 'r', linewidth=2, alpha=.3)
            hax[3,2].set_ylabel('FZ[deg]', fontsize=12)
            hax[3,0].set_xlabel('Frames'), hax[3,1].set_xlabel('Frames'), hax[3,2].set_xlabel('Frames')
            #plt.tight_layout()
            #plt.grid()
            plt.show()
       
        # Storing the data in a new variable for each subject and each velocity    
        # If the subject landed with his right foot, make this, left foot
        if s==0 and v==0 or s==1 and v==0 or s==1 and v==1 or s==1 and v==2 or s==2 and v==1 or s==2 and v==2 or s==3 and v==0 or s==3 and v==2 or s==3 and v==3 or s==6 and v==0 or s==7 and v==0 or s==7 and v==2:
            newdata = np.array([plvtilt2, plvlist2, plvrot2, hipflxR2, hipaddR2, hiprotR2, kneeR2, ankleR2, footR2, hipflxL2, hipaddL2, hiprotL2, kneeL2, ankleL2, footL2])
        else:
            newdata = np.array([plvtilt2, plvlist2, plvrot2, hipflxL2, hipaddL2, hiprotL2, kneeL2, ankleL2, footL2, hipflxR2, hipaddR2, hiprotR2, kneeR2, ankleR2, footR2])
        newdata2 = transpose(newdata)    
        dic['sub' + str(s) + 'vel' + str(v)] = newdata2 
        #print(dic)
#plt.plot(dic["sub0vel1"])
#print(dic["sub0vel1"].shape)
```

```python
# Calculating mean through all subjects in each velocity
d = {}
for v, var in enumerate(variable):                                          
    for vl, vel in enumerate(velocity):
        #fast
        d[var + '_' + vel] = np.mean([dic['sub0vel'+ str(vl)][:,str(v)],dic['sub1vel'+ str(vl)][:,str(v)],dic['sub2vel'+ str(vl)][:,str(v)],dic['sub3vel'+ str(vl)][:,str(v)],dic['sub4vel'+ str(vl)][:,str(v)],dic['sub5vel'+ str(vl)][:,str(v)],dic['sub6vel'+ str(vl)][:,str(v)],dic['sub7vel'+ str(vl)][:,str(v)]],axis=0)
#print (d)   
```
